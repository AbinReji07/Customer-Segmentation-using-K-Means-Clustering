```python
import pandas as pd
df=pd.read_csv("/content/Mall_Customers.csv")
df
```

```python
df.isna().sum()
```

```python
X=df.iloc[:,3:]
X
```

```python
from sklearn.cluster import KMeans
kmn=KMeans(n_clusters=4,init="k-means++",random_state=1)
```

```python
y=kmn.fit_predict(X)
y
```

```python
df["cluster"]=y
df
```

```python
#display the data points belons to cluster 1
df.loc[df["cluster"]==1]
```

```python
#elbow method-to find the optimal number of clusters
#it is a graph b/w wcss and number of clusters
#the first bend from the right side is taken as the optimal value for k
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init="k-means++",random_state=1)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

print(wcss)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss)
plt.show()
```
